```python
#getter is used to get the access of private or protected variable
# setter is used to set the private or protected variable
class c1:
    def __init__(self,name,section):
        self.__name=name
        self.section=section
    @property
    def getter(self):
        return self.__name
    @getter.setter
    def setter(self,new):
        self.__name=new
   
    @getter.deleter
    def delete(self):
        del self.__name
obj=c1('harsh',434344)
obj.section
print(obj.getter)
obj.setter='agrawal'
print(obj.getter)
del obj.delete
```

```python
# method overriding means that same function perform different function.

class Parent():
    def __init__(self):
        self.value = "Inside Parent"
    def show(self):
        print(self.value)
class Child(Parent):
    def __init__(self):
        self.value = "Inside Child"
    def show(self):
        print(self.value)
obj1 = Parent()
obj2 = Child()
  
obj1.show()
obj2.show()
```
